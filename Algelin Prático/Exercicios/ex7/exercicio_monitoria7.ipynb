##  Matrizes úteis para referência

* **Reflexão sobre o eixo $x$:**
  
 $ R_x = $
  $\begin{bmatrix}
  1 & 0 \\
  0 & -1
  \end{bmatrix}
  $

* **Reflexão na reta $y = x$:**
 $
  R_{y=x} = $
  $\begin{bmatrix}
  0 & 1 \\
  1 & 0
  \end{bmatrix}
  $

* **Rotação de $\theta$ graus anti-horária:**
 $
  R_\theta =$
 $ \begin{bmatrix}
  \cos \theta & -\sin \theta \\
  \sin \theta & \cos \theta
  \end{bmatrix}
  $

* **Translação por vetor $\vec{v} = (a, b)$:**
  $
  T(\vec{x}) = \vec{x} + \vec{v}
  $



```python
import numpy as np
import matplotlib.pyplot as plt

# função para plotar um triângulo
def plot_triangulo(T_transf, title):
    plt.figure(figsize=(3,3))
    tr = np.hstack((T_transf, T_transf[:,0:1]))
    plt.plot(tr[0,:], tr[1,:], marker='o')
    plt.title(title)
    plt.grid(True)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()

# triângulo dado
T = np.array([[1, 3, 2],
              [1, 1, 3]], dtype=float)

# reflexão sobre o eixo x
R_x = np.array([[1, 0],
                [0, -1]], dtype=float)

# reflexão sobre a reta x = y
R_xy = np.array([[0, 1],
                 [1, 0]], dtype=float)

# rotação de theta graus antihorário
def R_theta(theta_deg):
    th = np.deg2rad(theta_deg)
    return np.array([[np.cos(th), -np.sin(th)],
                     [np.sin(th),  np.cos(th)]], dtype=float)

v = np.array([[2.0], [-1.0]])

# a) Reflexão no eixo x
T_rx = R_x @ T
plot_triangulo(T_rx, "Reflexão no eixo x")

# b) Reflexão na reta x=y
T_rxy = R_xy @ T
plot_triangulo(T_rxy, "Reflexão na reta y = x")

# c) rotação por θ=45°
theta_example = 45.0
R45 = R_theta(theta_example)
T_rot45 = R45 @ T
plot_triangulo(T_rot45, f"Rotação anti-horária θ={theta_example}°")

# d) translação pelo vetor (2, -1)
T_trans = T + v
plot_triangulo(T_trans, "Translação por (2, -1)")




# Aplicações Diretas - Composição de Isometrias

# a) Matrix composta Rx * R90
R90 = R_theta(90.0)
M = R_x @ R90   

# b) aplicar sobre o triângulo
T_comp = M @ T

# c) plotar
plot_triangulo(T_comp, "Composição: rot 90° -> reflexão eixo x")

# repetir com theta = 45
M2 = R_x @ R45
T_comp2 = M2 @ T
plot_triangulo(T_comp2, "Composição: rot 45° -> reflexão eixo x")
```
